```python
def valid_moves(state,player):  # returns a list of all of the possible moves given a state *** means off the board need to account fr this
    moves=[]
    right_edge_index= [2,5,8]
    upper_edge_index= [0,1,2]
    
    #player 1 (up, down, right)
    if player == 1:
        #start: index 0
        if state[0]==1 and state[1]==0 and state[1] !=2:        #right
            moves.append([0,1])
        if state[0]==1 and state[3]==0 and state[3] !=2:        #down
            moves.append([0,3])
            
        #index 1
        if state[1]==1 and state[4]==0 and state[4] !=2:         #down
            moves.append([1,4])
        if state[1]==1 and state[2]==0 and state[2] !=2:         #right
            moves.append([1,2])
        
        #index 2*****
        if state[2]==1 and state[5]==0 and state[5] !=2:         #down
            moves.append([2,5])
        if state[2]==1:
            moves.append([2,2]) 
        
        #index 3
        if state[3]==1 and state[4]==0 and state[4] !=2:         #right
            moves.append([3,4])
        if state[3]==1 and state[0]==0 and state[0] !=2:         #up
            moves.append([3,0])
        if state[3]==1 and state[6]==0 and state[6] !=2:         #down
            moves.append([3,6])
            
        #index 4
        if state[4]==1 and state[5]==0 and state[5] !=2:         #right
            moves.append([4,5])
        if state[4]==1 and state[1]==0 and state[1] !=2:         #up
            moves.append([4,1])
        if state[4]==1 and state[7]==0 and state[7] !=2:         #down
            moves.append([4,7])
        
        #index 5****
        if state[5]==1 and state[2]==0 and state[2] !=2:         #up
            moves.append([5,2])
        if state[5]==1 and state[8]==0 and state[8] !=2:         #down
            moves.append([5,8])
        if state[5]==1:
            moves.append([5,5]) 
            
        #index 6
        if state[6]==1 and state[7]==0 and state[7] !=2:         #right
            moves.append([6,7])
        if state[6]==1 and state[3]==0 and state[1] !=2:         #up
            moves.append([6,3])
            
        #index 7
        if state[7]==1 and state[8]==0 and state[8] !=2:         #right
            moves.append([7,8])
        if state[7]==1 and state[4]==0 and state[4] !=2:         #up
            moves.append([7,4])
        
        #index 8******
        if state[8]==1 and state[5]==0 and state[5] !=2:         #up
            moves.append([8,5])
        if state[8]==1:
            moves.append([8,8]) 
        
    #player 2 (up, right, left)
    if player == 2:
         #start: index 0 *** 
        if state[0]==2 and state[1]==0:        #right
            moves.append([0,1])
        if state[0]==2:
            moves.append([0,0]) 
            
        #index 1 ***
        if state[1]==2 and state[0]==0:         #left
            moves.append([1,0])
        if state[1]==2 and state[2]==0:         #right
            moves.append([1,2])
        if state[1]==2:
            moves.append([1,1]) 
            
        #index 2**
        if state[2]==2 and state[1]==0:         #left
            moves.append([2,1])
        if state[2]==2:
            moves.append([2,2]) 
            
        #index 3
        if state[3]==2 and state[0]==0:         #up
            moves.append([3,0])
        if state[3]==2 and state[4]==0:         #right
            moves.append([3,3])
            
        #index 4
        if state[4]==2 and state[3]==0:         #left
            moves.append([4,3])
        if state[4]==2 and state[5]==0:         #right
            moves.append([4,5])
        if state[4]==2 and state[1]==0:         #up
            moves.append([4,1])
        
        #index 5
        if state[5]==2 and state[4]==0:         #left
            moves.append([5,4])
        if state[5]==2 and state[2]==0:         #up
            moves.append([5,2])
        
        #index 6
        if state[6]==2 and state[7]==0:         #right
            moves.append([6,7])
        if state[6]==2 and state[3]==0:         #up
            moves.append([6,3])
            
        #index 7
        if state[7]==2 and state[8]==0:         #right
            moves.append([7,8])
        if state[7]==2 and state[4]==0:         #up
            moves.append([7,4])
        if state[7]==2 and state[6]==0:         #right
            moves.append([7,6])
        
        #index 8
        if state[8]==2 and state[7]==0:         #left
            moves.append([7,8])
        if state[8]==2 and state[5]==0:         #up
            moves.append([8,5])
    return moves
```

```python
def minvalue_example(current_state,player,depth=1):
    if player==1:
        other_player=2
    else:
        other_player=1

    # since win_status is called with a player and an updated state
    # the current state is really the updated state from the
    # other player's last move.

    tabs='\t'*depth
    print(tabs,"MIN value current state\n")
    s=current_state
    print('\n'.join([tabs +_ for _ in str(s).split('\n')]))    
    
    status=win_status(current_state,other_player)

    if status=='win':  # bad for min
        print(tabs,"end state value: 1")
        return 1
    elif status=='lose':  # good for min
        print(tabs,"end state value: -1")
        return -1
    elif status=='stalemate':  # draw
        print(tabs,"end state value: 0")
        return 0


    moves=valid_moves(current_state,player)
    available_states=[update_state(deepcopy(current_state),player,move)
                                for move in moves]
    
    
    print(tabs,"Getting MAX value for states")
    for s in available_states:
        print('\n'.join([tabs +_ for _ in str(s).split('\n')]))    
    
    
    values=[]
    for state in available_states:
        value=maxvalue_example(state,other_player,depth+1)
        values.append(value)
                
    if not values:  # no available states  = stalemate
        return 0
    else:
        return min(values)
     
        
def maxvalue_example(current_state,player,depth=1):
    if player==1:
        other_player=2
    else:
        other_player=1

    # since win_status is called with a player and an updated state
    # the current state is really the updated state from the
    # other player's last move.
    tabs='\t'*depth
    print(tabs,"MAX value current state\n")
    s=current_state
    print('\n'.join([tabs +_ for _ in str(s).split('\n')]))    


    status=win_status(current_state,other_player)

    if status=='win':  # bad for max
        print(tabs,"end state value: -1")
        return -1
    elif status=='lose':  # good for max
        print(tabs,"end state value: 1")
        return 1
    elif status=='stalemate':  # draw
        print(tabs,"end state value: 0")
        return 0


    moves=valid_moves(current_state,player)
    available_states=[update_state(deepcopy(current_state),player,move)
                                for move in moves]
    print(tabs,"Getting MIN value for states")
    for s in available_states:
        print('\n'.join([tabs +_ for _ in str(s).split('\n')]))    
    
    values=[]
    for state in available_states:
        value=minvalue(state,other_player,depth+1)
        values.append(value)
                
    if not values:  # no available states  = stalemate
        return 0
    else:
        return max(values)
```

```python
def update_state(state,player,move):
    new_state=state
    if move[0]==move[1]:
        state[move[1]]=0
    else:
        try:
            state[move[1]]=player
            state[move[0]]=0
        except IndexError:
            print(state)
            print(move)
            print(player)
            raise
    return new_state
```

```python
def skittles_move(state,player,info):
    S=info.S
    last_action=info.last_action
    last_state=info.last_state
    
    
    # if Ive never seen this state before
    if not state in S:
        actions=valid_moves(state,player)

        S[state]=Table()
        for action in actions:
            S[state][action]=3     
    
    try:
        move=weighted_choice(S[state])  # weighted across actions
    except IndexError:
        print(state)
        #print(move)
        print(player)
        raise
    
    # what if there are no skittles for a particular state?
    # move is None in that case
    
    if move is None:
        # learn a little bit
        if last_state:
            S[last_state][last_action]=S[last_state][last_action]-1
            if S[last_state][last_action]<0:
                S[last_state][last_action]=0
        
        move=random_move(state,player)
    
    return move

def skittles_after(status,player,info):
    S=info.S
    last_action=info.last_action
    last_state=info.last_state

    if status=='lose':
        # learn a little bit
        S[last_state][last_action]=S[last_state][last_action]-1
        if S[last_state][last_action]<0:
            S[last_state][last_action]=0
        
    


skittles_agent=Agent(skittles_move)
skittles_agent.S=Table()
skittles_agent.post=skittles_after
```
